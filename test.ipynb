```python
# create a new df with the date and cases
df1 = df[['Date','Cases']]
# if Cases is NaN set it to 0 
df1.loc[df1['Cases'].isnull(),'Cases'] = 0
# for each date find the mean of cases
df1 = df1.groupby('Date').sum()
# remove the 3 last rows
df1 = df1[:-3]

# plot the data
plt.plot(df1.index,df1['Cases'])
plt.xlabel('Date')
plt.ylabel('Cases')
plt.title('Cases per day')
plt.show()
```

```python
df2 = df[['Date','Deaths']]
# if Deaths is NaN set it to 0 
df2.loc[df2['Deaths'].isnull(),'Deaths'] = 0
# for each date find the mean of cases
df2 = df2.groupby('Date').sum()
# remove the 3 last rows
df2 = df2[:-3]
# plot df2
df2.plot()
# label in the x axis
plt.xlabel('Ημέρα')
# label in the y axis
plt.ylabel('Αριθμός Συνολικών Θανάτων')
```

```python
# from df1 get an array of the cases
cases = df1['Cases'].to_numpy()
# from df2 get an array of the deaths
deaths = df2['Deaths'].to_numpy()

morality = deaths/cases
# plot the morality
plt.plot(morality)
plt.show()
```

```python
df3 = df[['Date','Cases']]
# if Deaths is NaN set it to 0 
df3.loc[df3['Cases'].isnull(),'Cases'] = 0
# for each date find the mean of cases
df3 = df3.groupby('Date').sum()
df3 = df3[:-3]
print(df3)
cases = df3['Cases'].to_numpy()

cases_per_date = [cases[0]]
for i in range(1,len(cases)):
    cases_per_date.append(cases[i]-cases[i-1])

dates = df3.index.to_numpy()

plt.plot(dates,cases_per_date)
```
